# **ANALISIS DE LA CALIDAD DEL AIRE DE MADRID AÑO 2018**

###**CONTEXTO**

Este dataset proviene de una recopilacion de datos de monitoreo en 24 estaciones en la ciudad de Madrid, España entre los años 2001-2018. La cual surge debido a la necesidad de implementar medidas de control, como restricciones de tráfico, ante episodios de alta contaminacion dada sobre todo durante estaciones secas. Estos daros fueron proporcionados originalmente por el Ayuntamiento de dicha ciudad y han sido puesto a disposicion libre, para cualquier estudio o analisis descriptivo.

No todas las estaciones miden todos los contaminantes
Por eso existen valores nulos en varias columnas

Los principales contaminantes incluidos son:

* SO_2: Dioxido de azufre medido en μg/m³.
* NO: Oxido nitrico medido en μg/m³.
* NO_2: Dioxido de nitrogeno  μg/m³.
* PM25: Particulas en suspension de diametro menor a 2.5 μm, μg/m³.
* PM10: Particulas en suspension de diametro menor a 10 μm, μg/m³
* NOx: Oxidos de nitrosos , μg/m³.
* O_3: ozono trospoférico, μg/m³.
* TOL: tolueno (methylbenzene) ; μg/m³.
* BEN: benzeno, μg/m³.
* EBE: ethylbenzeno  μg/m³.
* MXY: m-xyleno, μg/m³.
* PXY: p-xyleno, μg/m³.
* OXY: o-xyleno, μg/m³.
* TCH: total hydrocarbonos, mg/m³.
* CH4: methano, mg/m³.
* NMHC: no-methano hydrocarbonos (volátiles organicos compouestos), mg/m³.

## **CRUCE DE INFORMACION DE MAXIMOS CON LA FECHA DEL ELVENTO**

```python
df_2018 = df_2018.reset_index()
cols = [
    "BEN","CH4","CO","EBE","NMHC","NO","NO_2","NOx",
    "O_3","PM10","PM25","SO_2","TCH","TOL"
]

resultados = []

for col in cols:
    idx = df_2018[col].idxmax()
    fila = df_2018.loc[idx]

    resultados.append({
        "contaminante": col,
        "station": fila["station"],
        "fecha": fila["date"],
        "valor": fila[col]
    })

df_eventos = pd.DataFrame(resultados)
```

```python
df_eventos.sort_values("fecha")
```

Del resultado de eventos máximos se puede observar, que las maximas para PM10 y PM2,5 ocurren en disntintas fechas y horarios nocturnos, aunque puede ser que para una misma fecha ambos contaminantes hayan bordeado el maximo que se indica.Se revisará más adelante.

Así mismo para  BEN, CO, NMHC se ovserva que ocurren en la misma fecha y hora , 00:00, econtrandose tambien el max de TOL en un día anterior y 5 horas antes de las 22:00, lo cual sugiere que para la misma fecha TOL podría haber bordeado su máximo.

Para los oxidos Nitrosos, NO, NO2 Y NOx, se observa que se comportaron similar alcanzadno maximo en la misma fecha y hora pra el caso de NO y NOx, las cuales además pertenecen a la misma estacion. En el caso de NO2 el maximo se confirmo en la estacion "056" un día  anterior y a la misma hora que los otros 2 gases. De la tabla general de datos se confirma que para la misma fecha los 3 gases bordearon los maximos indicados. Esto indica claramente que este evento está relacionado a una misma causa.

para el caso de O_3 , ozone, se observa que su máximo se registra a las 02:00 horas a fines de Abril. Muy distante de la maxima de CO que se registra a fines de Enero. No obstante,sabiendo que CO es precursor de O_3 por ser catalizador químico, es posible para una misma fecha Los valores de ambos bordeen sus maximos respectivos. Esto se confirma en la tabla de maximos globales, en donde para la estacion que registra el maxima de CO (3,2) el valor de ozone observado por la misma estacion es de 117, elevado, aunque no tan cerano a su maxima de 133 registrada en abril. El caso opuesto de la tabla global mencionada se observa que en la estacion donde se registra la máxima para ozone no se dispone de informacion para CO.

Para continuar el analisis filtraremos por fecha y hora, a fín de inferir si los picos pueden estar relacionados a un mismo evento, lo cual se complementará con investigación bibliografica. Este paso se realizará usando power Bi para complementar más de un tipo de herramienta de analisis y visualización.

## **NORMALIZACION DEL DATA SET DE CALIDAD DL AIRE, Z-SCORE**

Se realiza una normalizacion de los datos, teniendo en cuenta cada columna de contaminante así como su estación de monitoreo. De esta manera es mas fiable la comparación entre columnas.Se aplicará una normalizacion Z-score.

```python
df_2018 = df_2018.drop(columns=["level_0", "index"], errors="ignore")
cols = [
    "BEN","CH4","CO","EBE","NMHC","NO","NO_2","NOx",
    "O_3","PM10","PM25","SO_2","TCH","TOL"
]
df_2018[cols] = df_2018.groupby("station")[cols].transform(
    lambda x: (x - x.mean()) / x.std()
)
```
